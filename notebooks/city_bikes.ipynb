# CityBikes

Send a request to CityBikes for the city of your choice. 

```python
import json
import os
import requests
from pprint import pprint
import pandas as pd
```

```python
network_id = 'belfastbikes-belfast'
url = f"http://api.citybik.es/v2/networks/{network_id}"
response = requests.get(url)
networks = response.json()['network']['stations']
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
len(networks)
networks[0]['latitude']
```

```python
latitude = []
longitude = []
empty_slots = []
free_bikes = []

for station in networks:
    
    latitude.append(station['latitude'])
    longitude.append(station['longitude'])
    empty_slots.append(station['empty_slots'])
    free_bikes.append(station['free_bikes'])

#creating a dictionary
bike_dict = {
    'latitude': latitude,
    'longitude': longitude,
    'empty_slots': empty_slots,
    'free_bikes': free_bikes
}
```

Put your parsed results into a DataFrame.

```python
bike_df = pd.DataFrame(bike_dict)
bike_df
```

```python
#Creating a new column that would show the total bikes a station can hold
bike_df['total_bike_slots'] = bike_df['empty_slots']+bike_df['free_bikes']
bike_df
```
